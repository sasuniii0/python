```python
import pandas as pd;

df = pd.DataFrame({
    "Product" : ["Laptop" , "Mouse" , "Keyboard" , "Monitor" , "Headset"],
    "UnitsSold" : [12,50,34,20,40],
    "Price" : [120000,1500,3500,25000,6000],
    "Discount" : [0.10,0.05,0.08,0.12,0.07]
})

df['FinalPrice'] = df['Price'] * (1 - df['Discount'])

df['TotalRevenue'] = df['UnitsSold'] * df["FinalPrice"]
print(df)

print("------------------------------------------------------------------------")

result = df[df["TotalRevenue"] > 500000]
print(result)

print("-------------------------------------------------------------------------")

all_total_revenue = df["TotalRevenue"].sum()
print(all_total_revenue)
```
