```python
class Car:
    wheels = 4

    def __init__(self, color, brand):
        self.brand = brand
        self.color = color

    def drive_to(self, destination):
        return f"The {self.color} {self.brand} is driving to {destination}"

    def show_wheel_count(self):
        return f"The {self.color} {self.brand} runs on {self.wheels} wheels"

car_1 = Car("red", "BMW")
print(car_1.drive_to("Colombo"))
print(car_1.show_wheel_count())
```

```python
from math import pi 
class Circle:
    pi = pi

    def __init__(self, radius):
        self.radius = radius
    
    def circumference(self):
        return 2 * self.pi * self.radius

    def area(self):
        return self.pi * (self.radius ** 2)
    
    def display_info(self):
        return f"The Circle's Radius: {self.radius}, Area: {self.area()}, Circumference: {self.circumference()}"
    

circle_1 = Circle(5)

print(circle_1.area())
print(circle_1.circumference())
print(circle_1.display_info())
```

```python
class Animal:

    def __init__(self, name):
        self.name = name 

    def speak(self):
        return f"{self.name} speak!"

class Dog(Animal):

    def __init__(self, name, color):
        super().__init__(name)
        self.color = color

    def speak(self):
        return f"The {self.color} color {self.name} barks!"

dog_1 = Dog("Tommy","Brown")
print(dog_1.speak())
```
