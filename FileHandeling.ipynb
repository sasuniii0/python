```python
file_1 = open('file_1.txt', 'r')
content = file_1.read()
print(content)
file_1.close()
```

```python
with open('file_1.txt', 'r') as file_1:
    content = file_1.read()
    print(content)
```

```python
with open('file_1.txt', 'r') as file_1:
    line_1 = file_1.readline()
    print(line_1)

    line_2 = file_1.readline()
    print(line_2)
```

```python
with open('file_1.txt', 'r') as file_1:
    line_1 = file_1.readline()
    print(line_1.strip())

    line_2 = file_1.readline()
    print(line_2)
```

```python
with open('file_1.txt', 'r') as file_1:
    full_content = file_1.readline()
    while full_content:
        print(full_content.strip())
        full_content = file_1.readline()
        
```

```python
with open('file_1.txt', 'r') as file_1:
    while True:
        full_content = file_1.readline()
        if not full_content:
            break
        print(full_content.strip()) 
```

```python
with open('file_1.txt', 'r') as file_1:
    content = file_1.readlines()
    print(content)
```

```python
with open('file_1.txt', 'r') as file_1:
    content = file_1.readlines()
    for line in content:
        print(line.strip())
```
