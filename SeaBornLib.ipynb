```python
sns.set_style('whitegrid')
sns.set_style("darkgrid")
sns.set_style('ticks')
sns.scatterplot(data = penguins , x = "flipper_length_mm" , y = "body_mass_g", hue="sex")
```

```python
sns.set_style('ticks')
sns.set_context("paper") # idol
sns.scatterplot(data = penguins , x = "flipper_length_mm" , y = "body_mass_g", hue="sex")
sns.despine(left=True)
```

```python
sns.set_style('ticks')
sns.set_context("notebook") 
sns.scatterplot(data = penguins , x = "flipper_length_mm" , y = "body_mass_g", hue="sex" , palette='Set1')
sns.despine(left=True)
```

```python
columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
sns.pairplot(data = penguins[columns])
```

```python
graph = sns.PairGrid(data = penguins, hue="sex", palette="Set2")
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
```
