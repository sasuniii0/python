```python
import csv 

# Define the CSV file path
csv_file_path = 'sample_output.csv'

# Sample data to write to CSV
data =[
    {'Name': 'Alice', 'Age': 30, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 25, 'City': 'Los Angeles'},
    {'Name': 'Charlie', 'Age': 35, 'City': 'Chicago'},
]

# Writing data to CSV using DictWriter
with open(csv_file_path , 'w' , newline='') as csv_file:
    # Define the fieldnames
    fieldnames = ['Name', 'Age', 'City']

    # Create a DictWriter object
    csv_dict_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    # Write the header
    csv_dict_writer.writeheader()
    
    # Write the data rows
    csv_dict_writer.writerows(data)
```

```python
with open('input.txt' , 'r') as input_file:
    input_file_lines = input_file.readline()
    
    even_lines = []
    index = 0

    for line in even_lines:
        if index % 2 == 0:
            even_lines.append(line)
        index += 1
    with open('even_lines.txt', 'w') as even_file:
        even_file.writelines(even_lines)
```

```python
with open('input.txt', 'r') as input_file:
    input_file_lines = input_file.readlines()

    with open("even_lines.txt", 'w') as even_file:
        for i in range(1 , len(input_file_lines) - 1, 2):
            even_file.write(input_file_lines[i])
```
