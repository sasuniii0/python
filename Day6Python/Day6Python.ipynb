```python
class StudentRecords:
    def __init__(self, grades):
        self.__grade = list(grades)

    def add_grade(self, grade):
        if not self.__validate_grade(grade):
            return "invalid grade: must be number between 0 and 100"
        self.__grade.append(grade)
        return "added successfully"

    def __validate_grade(self, grade=None):
        if grade is not None:
            if not isinstance(grade, (int, float)):
                return False
            return 0 <= grade <= 100

        if not isinstance(self.__grade, (list, tuple)):
            return False
        for g in self.__grade:
            if not isinstance(g, (int, float)):
                return False
            if g < 0 or g > 100:
                return False
        return True

    def get_average(self):
        return self.__calculate_average()

    def __calculate_average(self):
        if not self.__validate_grade():
            raise ValueError("Grades must be numeric and between 0 and 100.")
        if len(self.__grade) == 0:
            raise ValueError("No grades available to calculate average.")
        return sum(self.__grade) / len(self.__grade)
```

```python
class StudentRecords:
    def __init__(self):
        self.__grade = []

    def add_grade(self,grade):
        if self.__validate_grade(grade):
            self.__grade.append(grade)
            return "added successfully"
        else:
            return "invalid grade: must be number between 0 and 100"
        
    def __validate_grade(self, mark):
        return isinstance(mark, int) and 0 <= mark <= 100
    
    def get_average(self):
            return self.__calculate_average()
    
    def __calculate_average(self):
        if not self.__grade:
            return 0
        return sum(self.__grade) / len(self.__grade)
        
student = StudentRecords()
print(student.add_grade(95))
print(student.add_grade(88))
print(student.get_average())
```
