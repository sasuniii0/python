```python
import numpy as np;

array_1 = np.array([1,2,3,4,5])
array_2 = np.array([6,7,8,9,10])

sum_of_array = array_1 + array_2
print(sum_of_array)

difference_of_array = array_1 - array_2
print(difference_of_array)

mult_of_array = array_1 * array_2
print(mult_of_array)

dev_of_array = array_1/ array_2
print(dev_of_array)
```

```python
array = np.array([1,2,3,4,5,6])

#Prints sum of the array
summaration = np.sum(array)
print(summaration)

#Prints average
average = np.mean(array)
print(average)

#Print max of the array
print(np.max(array))

#Print min of the array
print(np.min(array))

#Print sqrt of the array each elements0
print(np.sqrt(array))
```
