```python
def sum_of_squares(n):

    if n < 1 :
        return "Input should be a positive number"
    
    total = 0
    for i in range(1, n + 1):
        total += i ** 2
    return total

sum_of_squares(2)
```

```python
def add_corresponding_numbers(list_1, list_2):

    if len(list_1) != len(list_2):
        return "Lists are not equal of length"

    new_list = []
    for i in range(len(list_1)):
        new_list.append(list_1[i] + list_2[i])

    return new_list
print(add_corresponding_numbers([1, 2, 3], [4, 5, 6]))
```

```python
def average_positive(*numbers):
    positive_numbers = []

    for num in numbers:
        if num > 0:
            positive_numbers.append(num)

    if len(positive_numbers) == 0:
        return "No positive numbers provided"

    average = sum(positive_numbers) / len(positive_numbers)
    return average

print(average_positive(2, 2, 2, 2, 2))
print(average_positive(1, 2, 3,-2,-5))
print(average_positive(-1, -4, -5))
```

```python
def average_positive(*numbers):
    total = 0
    count = 0

    for num in numbers:
        if num > 0:
            total += num
            count += 1

    if count == 0:
        return "No positive numbers provided"
        
    return total/ count
print(average_positive(1,3,4,5,6))
print(average_positive(-1,4,6,-4,7,-8))
```

```python
def average_score(**subjects):
    if not subjects:  
        return "No Subject Provided"

    total = 0

    for score in subjects.values():
        total += score

    average = total / len(subjects) 
    return average

print(average_score(maths=76, physics=90))
print(average_score(it = 90, statics = 89, chemistry = 88))
```

```python
def fahrenheit_to_celsius(fahrenheit_list):
    def convert_to_celsius(fahrenheit):
        celcius = (fahrenheit - 32) * 5/9
        celcius = round(celcius , 2)
        return celcius
    
    result = list(map(convert_to_celsius,fahrenheit_list))
    return result

fahrenheit_list = [23,56,87,45,35]
fahrenheit_to_celsius(fahrenheit_list)
```

```python
def filter_even_numbers(numbers):
    if numbers % 2 == 0:
        return True
    else: 
        return False
    
numbers = [9,5,8,12,13]
result = list(filter(filter_even_numbers,numbers))
print(result)
```

```python
def extract_numbers(n):
    return n % 3 == 0 and n % 5 != 0

numbers = [12, 15, 18, 20, 25, 30, 35]
result = filter(extract_numbers, numbers)

print(list(result))
```
