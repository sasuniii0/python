```python
def sum_of_squares(n):

    if n < 1 :
        return "Input should be a positive number"
    
    total = 0
    for i in range(1, n + 1):
        total += i ** 2
    return total

sum_of_squares(2)
```

```python
def add_corresponding_numbers(list_1, list_2):

    if len(list_1) != len(list_2):
        return "Lists are not equal of length"

    new_list = []
    for i in range(len(list_1)):
        new_list.append(list_1[i] + list_2[i])

    return new_list
print(add_corresponding_numbers([1, 2, 3], [4, 5, 6]))
```

```python
def calculate_total(price, quantity = 4, discount = 0):
    total = price * quantity
    discount_cal = total * (discount / 100)
    return total - discount_cal

print(calculate_total(250))
print(calculate_total(300, 55))
print(calculate_total(550, 20, 7))
```

```python
def sum_numbers(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(sum_numbers(1, 2, 3, 4))
print(sum_numbers(1, 2))
print(sum_numbers())
```

```python
def average_positive(*numbers):
    positive_numbers = []

    for num in numbers:
        if num > 0:
            positive_numbers.append(num)

    if len(positive_numbers) == 0:
        return "No positive numbers provided"

    average = sum(positive_numbers) / len(positive_numbers)
    return average

print(average_positive(2, 2, 2, 2, 2))
print(average_positive(1, 2, 3,-2,-5))
print(average_positive(-1, -4, -5))
```

```python
def average_positive(*numbers):
    total = 0
    count = 0

    for num in numbers:
        if num > 0:
            total += num
            count += 1

    if count == 0:
        return "No positive numbers provided"
        
    return total/ count
print(average_positive(1,3,4,5,6))
print(average_positive(-1,4,6,-4,7,-8))
```

```python
def arbitrary_keyword_args(**kwargs):
    for key in kwargs.keys():
        print(key)

    for value in kwargs.values():
        print(value)
        
arbitrary_keyword_args(name = "John" ,address = "Colombo", age = 21)
```
