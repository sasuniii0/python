```python
total = 0
for i in range (10 ,0 ,-2):
    total += 1
print(total)
```

```python
result = ""
for i in range (3):
    for j in range(i):
        result += f"({i}, {j})"
print(result)
```

```python
count = 5
total = 0
while True:
    total += count
    count -= 1
    if count == 0:
        break
print(total)
```

```python
def process_data(**kwargs):
    status = kwargs.pop('status' , 'Pending')
    return f"Status: {status}, Remaining: {len(kwargs)}"

print(process_data(id = 101, name = "Alice", status = "Completed"))
```

```python
data = "The 1 quick 2 brown 3 fox"
filtered_list = [w.upper() for w in data.split() if w.isalpha()]
print(filtered_list)
```

```python
matrix = [[1,2] , [3,4]]
flat_list = [num * 2 for row in matrix for num in row]
print(flat_list)
```

```python
words = [ "apple", "banana", "cherry" ]
result = list(map(lambda x : x[0] , words))
print(result)
```
